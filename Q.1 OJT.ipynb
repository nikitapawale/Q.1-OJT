```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("iris.csv")
df.sample(5)
```

```python
#Histrog 
df=df.iloc[:,1:]
df.hist(figsize=(16,8))
plt.suptitle("Histrogram of all columns in iris dataset")
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(10,8))
sns.pairplot(df)
plt.suptitle("Scatter plot for all numerical colums")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df)
plt.suptitle("Boxplot for all numerical colums")
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Sepal.Length'])

plt.subplot(2,2,2)
sns.distplot(df['Sepal.Width'])

plt.subplot(2,2,3)
sns.distplot(df['Petal.Length'])

plt.subplot(2,2,4)
sns.distplot(df['Petal.Width'])
```
